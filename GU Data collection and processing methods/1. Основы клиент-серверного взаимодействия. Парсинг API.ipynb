#### Задание 1.  
Посмотреть документацию к API GitHub, разобраться как вывести список репозиториев для конкретного пользователя, сохранить JSON-вывод в файле *.json.  

```python
import requests
import json
from pprint import pprint
from IPython.display import Image
```

```python
name = 'Christinayar'

main_link = f'https://api.github.com/users/{name}/repos'
```

```python
def show_repository():
    req = requests.get(main_link)
    data = req.json()
    print(*[i['html_url'] for i in data], sep='\n')
```

```python
show_repository()
```

```python
def download_repos():
    req = requests.get(main_link)
    data = req.json()
    n = [i['html_url'] for i in data]
    
    with open('repositories.json', 'w') as f:
        json.dump(n, f)
```

```python
download_repos()
```
