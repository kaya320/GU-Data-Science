## №1. 

Нарисуйте график функции: y(x) = k∙cos(x – a) + b для некоторых (2-3 различных) значений параметров k, a, b

```python
k = .5
a = 1
b = 0
x = np.linspace(-np.pi, 3 * np.pi, 201)
plt.plot(x, np.cos(x))
plt.plot(x, k * np.cos(x - a) + b)
plt.xlabel('x')
plt.ylabel('y')
plt.axis('tight')
plt.grid(True)
plt.show()
```

## №2. 

Напишите код, который будет переводить полярные координаты в декартовы.

```python
phi = np.linspace(0, 2 * np.pi, 1000)
rho = 10
x = rho * np.cos(phi)
y = rho * np.sin(phi)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

Напишите код, который будет рисовать график окружности в полярных координатах.

```python
x = np.linspace(0, 15, 100)
y = 2*x
plt.polar(x, y)
```

```python
t = np.linspace(0, np.pi, 1000)
x = np.cos(t)
y = np.sin(t)

rho = np.sqrt(x ** 2 + y ** 2)
phi = np.arctan2(y, x) * 180 / np.pi
plt.polar(phi, rho)
```

```python
def xy2pol(x, y): 
    rho = np.sqrt(x ** 2 + y ** 2)
    phi = np.arctan2(y, x) * 180 / np.pi
    return(rho, phi)
    
t = np.linspace(0, np.pi, 1000)
x = np.cos(t)
y = np.sin(t)
rho, phi = xy2pol(x, y)
plt.polar(phi, rho)
```

Напишите код, который будет рисовать график отрезка прямой линии в полярных координатах.

```python
t = np.linspace(0, np.pi, 1000)
x = np.sin(t)
y = np.sin(t)
rho = np.sqrt(x ** 2 + y ** 2)
phi = np.arctan2(y, x) * 180 / np.pi
plt.polar(phi, rho)
```

```python
# Система уравнений
plt.figure(figsize=(10, 10))
x = np.linspace(-6, 6, 1001)
ind = np.where(x == 0.)
x = np.delete(x, ind)
y1 = x ** 2 - 1
y2 = (np.exp(x) - 1) / x + 1
plt.plot(x, y1, label = 'x ** 2 - 1', c='g')
plt.plot(x, y2, label = 'e ** 2 - 1 / x +1', c='b')
plt.axis([-6, 6, -3, 30])
plt.legend(loc = 'upper left')
plt.grid()
plt.show()
```

```python
from scipy.optimize import fsolve

def equations(p):
    x, y = p
    return (y - x ** 2 + 1, np.exp(x) + x * (1 - y) - 1)

x_1, y_1 = fsolve(equations, (-2, 5))
x_2, y_2 = fsolve(equations, (3, 6))
x_3, y_3 = fsolve(equations, (5, 20))

print(x_1, y_1)
print(x_2, y_2)
print(x_3, y_3)
```
