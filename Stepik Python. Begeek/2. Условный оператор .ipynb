Если пользователь захочет поделить на ноль, выведите текст «На ноль делить нельзя!».

На вход программе подаются два целых числа, каждое на отдельной строке, и строка. Программа должна вывести результат применения операции к введенным числам или соответствующий текст, если операция неверная либо если происходит деление на ноль.

```python
a, b = int(input()), int(input())
st = input()
if st == '+':
    print(a + b)
elif st == '-':
    print(a - b)
elif st == '*':
    print(a * b)
elif st == '/':
    if b == 0:
        print('На ноль делить нельзя!')
    else:
        print(a / b)
else:
    print('Неверная операция')
```
