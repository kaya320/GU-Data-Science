## №3. 

Напишите код на Python, реализующий построение графиков: окружности, эллипса, гиперболы.

**Парабола**

```python
x = []
y1 = []
y2 = []
p = 5
for i in range(1000):
    x_ = i / 1000
    x.append(x_)
    y1.append(math.sqrt(2 * p * x_))
    y2.append(-math.sqrt(2 * p * x_))
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
```

**Окружность**

```python
x = []
y1 = []
y2 = []
R = 2
for i in range(1000 * 2 * R + 1):
    x_ = -R + i/1000
    x.append(x_)
    y1.append(math.sqrt(R ** 2 - x_ ** 2))
    y2.append(-math.sqrt(R ** 2 - x_ ** 2))
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-3, 3)
plt.xlim(-4.3, 4.3)
plt.grid(True)
```

**Эллипс**

```python
x = []
y1 = []
y2 = []
a = 4
b = 2
for i in range(1000 * 2 * a + 1):
    x_ = -a + i / 1000
    x.append(x_)
    y1.append(math.sqrt(1 - x_ ** 2 / a ** 2 ) * b)
    y2.append(-math.sqrt(1 - x_ ** 2 / a ** 2 ) * b)
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-3, 3)
plt.ylim(-4.3, 4.3)
plt.grid(True)
```

**Гипербола**

```python
x = []
y1 = []
y2 = []
a = 1
b = 3
for i in range(1000 * 2 * a):
    x_ = a + i / 1000
    x.append(x_)
    y1.append(math.sqrt(x_ ** 2 / a ** 2 - 1) * b)
    y2.append(-math.sqrt(x_ ** 2 / a ** 2 -1 ) * b)
plt.plot(x, y1)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-3, 3)
plt.ylim(-3.3)
plt.grid(True)
```

Нарисуйте трехмерный график двух параллельных плоскостей.
Нарисуйте трехмерный график двух любых поверхностей второго порядка.

```python
from pylab import *

fig = figure()
ax = Axes3D(fig)

X = np.arange(-5, 5, 1)
Y = np.arange(-5, 5, 1)
X, Y = np.meshgrid(X, Y)
Z = 2 * X + 3 * Y 
Z2 = 2 * X + 3 * Y + 30 
ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y, Z2)
show()
```

```python
x = np.outer(np.ones(30,), np.linspace(-4, 4, 30))
y = x.copy().T

z = x ** 2 + y ** 2
fig = plt.figure()
ax = plt.axes(projection = '3d')

ax.plot_surface(x, y, z, cmap = plt.cm.jet)
show()
```
